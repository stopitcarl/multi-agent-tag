```python
width = np.pi * 2 / N
width
```

```python
colors = plt.cm.viridis(radii / 10.)
colors
```

```python
ax = plt.subplot(projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.5)
plt.show()
```

```python
def make_bar_plot(values : np.array):
    # Compute pie slices
    N = len(values)
    theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
    width = np.pi * 2 / N
    colors = [plt.cm.viridis(0.6)]
    ax = plt.subplot(projection='polar')
    ax.bar(theta, values, width=width, bottom=0.0, color=colors, alpha=0.5)
    plt.show()
```

```python
N = 21
mu = 0
sigma = N/12
x = np.linspace(-N/2, N/2, N)
y = stats.norm.pdf(x, mu, sigma)*60
plt.plot(x, y)
plt.show()
```

```python

mu = 0
sigma = N/40
x = np.linspace(-N/2, N/2, N)
y2 = stats.norm.pdf(x, mu, sigma)*60
plt.plot(x, y2)
plt.show()
```

```python
def add_danger(current_danger, new_danger, angle):
    N = len(current_danger)
    offset = (N - int(angle * N/(2*np.pi) - N/2)) % N
    print(offset)
    danger = current_danger - np.concatenate((new_danger[offset:],new_danger[:offset]))
    return danger
```

```python
ones = np.ones(N) * max(y2)
```

```python
danger = add_danger(ones, y, np.pi/2)
danger = add_danger(danger, y2, -np.pi/4)
make_bar_plot(danger)
```
